```python
import matplotlib.pyplot as plt

# ‚úÖ Í∏∞Î≥∏ ÏûÖÎ†•Í∞í
monthly_invest = 500_000   # Îß§Îã¨ Ìà¨Ïûê Í∏àÏï° (50Îßå Ïõê)
annual_return = 0.08       # ÏòàÏÉÅ Ïó∞ ÏàòÏùµÎ•† (8%)
years = 10                 # Ìà¨Ïûê Í∏∞Í∞Ñ (10ÎÖÑ)

# ‚úÖ Í≥ÑÏÇ∞
total_months = years * 12
monthly_return = (1 + annual_return) ** (1/12) - 1

balance = []
amount = 0

for month in range(total_months):
    amount = amount * (1 + monthly_return) + monthly_invest
    balance.append(amount)

# ‚úÖ Í≤∞Í≥º Ï∂úÎ†•
final_value = round(balance[-1], 2)
print(f"\nüìä {years}ÎÖÑ ÌõÑ ÏòàÏÉÅ ÏûêÏÇ∞: {final_value:,.0f}Ïõê")

# ‚úÖ Í∑∏ÎûòÌîÑ ÏãúÍ∞ÅÌôî
plt.plot(balance)
plt.title("Ïû¨ÌÖåÌÅ¨ ÏûêÏÇ∞ ÏÑ±Ïû• Í∑∏ÎûòÌîÑ")
plt.xlabel("Í∞úÏõî Ïàò")
plt.ylabel("ÎàÑÏ†Å ÏûêÏÇ∞ (Ïõê)")
plt.grid(True)
plt.show()
```
